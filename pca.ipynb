```python
import pandas as pd
df=pd.read_csv('C:\\Users\\neeth\\Downloads\\mlLab\\dataset\\iris(For PCA Program).csv')
```

```python
x=df.values[:,0:4]
y=df.values[:,-1]
```

```python
xstd=x-x.mean()
ystd=y-y.mean()
```

```python
import numpy as np
from numpy import linalg
cov=np.cov(x.T)
print(cov)
```

```python
lambdas,vs=linalg.eig(cov)
sortedIndex=np.argsort(lambdas)[::-1]
sorted_eigenvalue=lambdas[sortedIndex]
sorted_eigenvectors = vs[:,sortedIndex]
```

```python
n_components=2
eigenvector_subset = sorted_eigenvectors[:,0:n_components]
```

```python
X_reduced = np.dot(eigenvector_subset.transpose(),x.transpose()).transpose()
print("Information/Variance in PC1",(sorted_eigenvalue[0]/(sorted_eigenvalue[0]+sorted_eigenvalue[1]))*100)
```

```python
import matplotlib.pyplot as plt
plt.xlabel('Principal Component - 1',fontsize=20)
plt.ylabel('Principal Component - 2',fontsize=20)
plt.scatter(X_reduced[:,0:1],X_reduced[:,-1],c=y)
plt.show()
```
